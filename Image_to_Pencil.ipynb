```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
from google.colab.patches import cv2_imshow

from PIL import Image, ImageFilter
import numpy as np
import cv2
```

```python
# Load the image
image_path = '/content/drive/MyDrive/Strata Projects/Image to Pencil/dog.jpg'  # Image path
image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Save the grayscale image
# gray_image.save('/content/drive/MyDrive/Strata Projects/Image to Pencil/dog_grayscale_image.jpg')
```

```python
# Invert the grayscale image
inverted_image = 255 - gray_image

# Save the inverted image
# inverted_image.save('/content/drive/MyDrive/Strata Projects/Image to Pencil/dog_inverted_image.jpg')
```

```python
# Blur the inverted image
blurred_inverted_image = cv2.GaussianBlur(inverted_image, (5, 5), 0)
display(blurred_inverted_image)
# Save the blurred inverted image
# blurred_inverted_image.save('/content/drive/MyDrive/Strata Projects/Image to Pencil/dog_blurred_inverted_image.jpg')
```

```python
# Convert images to NumPy arrays
gray_array = gray_image
blurred_inverted_array = blurred_inverted_image

# Avoid division by zero
blurred_inverted_array[blurred_inverted_array == 0] = 1

# Mix the images
pencil_sketch_array = gray_array / blurred_inverted_array * 255

# Clip values to the valid range
pencil_sketch_array = np.clip(pencil_sketch_array, 0, 255).astype(np.uint8)

# Create a new image from the resulting array
pencil_sketch_image = cv2.cvtColor(pencil_sketch_array, cv2.COLOR_GRAY2BGR)

# Save the pencil sketch
# pencil_sketch_image.save('/content/drive/MyDrive/Strata Projects/Image to Pencil/dog_pencil_sketch.jpg')
```
